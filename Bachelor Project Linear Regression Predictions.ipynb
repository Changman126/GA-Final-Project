```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Data
df = pd.read_csv('BDataF.csv')
df_test = pd.read_csv('BDataTest.csv')

# Dropping unwanted columns
#cols_to_drop = ['BachelorAge', 'HairColor', 'HairType', 'Hometown']
#test_cols_to_drop = ['BachelorAge']
#df.drop(cols_to_drop, axis=1, inplace=True)
#df_test.drop(test_cols_to_drop, axis=1, inplace=True)

# Renaming Dummy Columns
df = pd.concat([df, pd.get_dummies(df['DateRose'], prefix='DateRose')], axis=1)
df_test = pd.concat([df_test, pd.get_dummies(df_test['DateRose'], prefix='DateRose')], axis=1)


BachelorData = df
BachelorTestData = df_test
```

```python
# List column names for feature selection
list(BachelorData)
```

```python
# Linear Regression using scikit-learn
feature_cols = ['FirstImpressionRose', 'DateRose', 'OlderthanBachelor', 'YoungerthanBachelor', 'SameAgeasBachelor', 'DateRose_0', 'DateRose_1', 'DateRose_2']

X = BachelorData[feature_cols]
y = BachelorData['Winner']

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)

# Print the Intercepts and Coefficients
print lm.intercept_
print lm.coef_

# Pair feature names withcoefficients
coefficients = zip(feature_cols, lm.coef_)

# R Squared value
lm.score(X, y)

coefficients
```

```python
# Make predictions for feature_cols_test and store them/print them
feature_cols_test = ['FirstImpressionRose', 'DateRose', 'OlderthanBachelor', 'YoungerthanBachelor', 'SameAgeasBachelor', 'DateRose_0', 'DateRose_1', 'DateRose_2']

X_new = BachelorTestData[feature_cols_test]
BachelorTestData['Predictions'] = lm.predict(X_new)

BachelorTestData
```
